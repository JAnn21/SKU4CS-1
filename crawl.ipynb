```python
from bs4 import BeautifulSoup
```

```python
import urllib.request

import pandas as pd
import datetime
from itertools import count
```

```python
def geturl(url, enc='utf-8'):
    
    req = urllib.request.Request(url)

    try:    
        
        response = urllib.request.urlopen(req)
        if response.getcode() == 200:
            try:                  
                rcv = response.read()
                ret = rcv.decode(enc)
            except UnicodeDecodeError:
                ret = rcv.decode(enc, 'replace')    
            
            return ret

    except Exception as e:
        print(e)
        print("[%s] Error for URL : %s" % (datetime.datetime.now(), url))
        return None
```

```python
def carresult(result):
   
    for page_idx in range(200,300):
        plist=[]
        kmlist=[]
        flist=[]        
        ylist=[]
        tlist=[]
        URL = 'https://www.bobaedream.co.kr/cyber/CyberCar.php?sel_m_gubun=ALL&page=%s&price_sel1=P0&order=S11&view_size=70' \
                       % str(page_idx + 1)
        print ("[Cars] : [%s]" % (str(page_idx + 1)))
        try:
            rcv_data = geturl(URL)
            soupData = BeautifulSoup(rcv_data, 'html.parser')
        
            products = soupData.findAll('li', attrs={'class':'product-item'})
            year = soupData.findAll('div',attrs={'class':'mode-cell year'})
            fuel = soupData.findAll('div',attrs={'class':'mode-cell fuel'})
            km = soupData.findAll('div',attrs={'class':'mode-cell km'})
            price = soupData.findAll('div',attrs={'class':'mode-cell price'})
        
        except:
            pass
        for i in range(1,len(price)):
            pp = price[i].find('strong')
            try:
                plist.append(str(pp).split('>')[1].split('<')[0]+"만원")
            except:
                plist.append('팔림')
   
        for i in range(1,len(year)):
            ylist.append(str(year[i]).split('"text">')[1].split('<')[0])
            kmlist.append(str(km[i]).split('"text">')[1].split('<')[0])
            flist.append(str(fuel[i]).split('"text">')[1].split('<')[0])
           
        for store_tr in products:
            bEnd = False
            bt = store_tr.find('div', attrs={'class':'mode-cell title'})
            title = bt.find('a').get('title')
            tlist.append(title)
        
        
        for i in range(len(tlist)):
            result.append([tlist[i]]+[ylist[i]]+[flist[i]]+[kmlist[i]]+[plist[i]])

    return
```

```python
def bobae():

    result = []
    carresult(result)
    print(result)
    carcsv = pd.DataFrame(result, columns=('name', 'year', 'fuel', 'km','price'))
    carcsv.to_csv("C:/Users/user/Desktop/cars2.csv", encoding="cp949",mode='w', index=True)
    print('FINISHED')
    
```
